```python
import numpy as np

df.groupby("Region").agg({"Units":(np.nansum,np.nanprod),
                                      "UnitCost":np.nansum})
```

```python
cols=['Item','UnitCost']
transform_df=df[cols].groupby('Item').transform(np.nanmean)
transform_df.head()
```

```python
transform_df.rename({'UnitCost':'UnitCostMean'}, axis='columns', inplace=True)
transform_df.head()
```

```python
df=df.merge(transform_df, left_index=True, right_index=True)
#Add this column to the original Dataframe (e.g. each Pencil has the same UnitCostMean!)
df.head()
```

```python
#calculate absulute difference between two columns
df['DiffToMeanCost']=np.absolute(df['UnitCost']-df['UnitCostMean'])
df['DiffToMeanCost'].head()
df.head()
```

```python
#only show region if average unit cost per region is more than 50 - this only applies to West
df.groupby('Region').filter(lambda x: np.nanmean(x['UnitCost'])>50)
```
