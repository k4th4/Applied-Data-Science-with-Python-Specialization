```python
import pandas as pd

staff_df = pd.DataFrame([{'First Name': 'Kelly', 'Last Name': 'Desjardins', 
                          'Role': 'Director of HR', 'Location': 'Washington Avenue'},
                         {'First Name': 'Sally', 'Last Name': 'Brooks', 
                          'Role': 'Course liasion', 'Location': 'Clinton Avenue'},
                         {'First Name': 'James', 'Last Name': 'Wilde', 
                          'Role': 'Grader', 'Location': 'Main Market'},
                         {'First Name': 'Adam', 'Last Name': 'Ni',
                          'Role': 'Researcher', 'Location': 'Office Builiding F'}])
student_df = pd.DataFrame([{'First Name': 'James', 'Last Name': 'Hammond', 
                            'School': 'Business', 'Location': '1024 Billiard Avenue'},
                           {'First Name': 'Mike', 'Last Name': 'Smith', 
                            'School': 'Law', 'Location': 'Homehouse 3'},
                           {'First Name': 'Sally', 'Last Name': 'Brooks', 
                            'School': 'Engineering', 'Location': 'Espenhain'},
                           {'First Name': 'Kelly', 'Last Name': 'Desjardins', 
                          'School': 'Finance', 'Location': 'Market Street'}])
```

```python
staff_df
```

```python
student_df
```

```python
pd.merge(staff_df, student_df, how='left', on=['First Name', 'Last Name'])
```

```python
pd.merge(staff_df, student_df, how='right', on=['First Name', 'Last Name'])
```

```python
pd.merge(staff_df, student_df, how='inner', on=['First Name', 'Last Name'])
```

```python
pd.merge(staff_df, student_df, how='outer', on=['First Name', 'Last Name'])
```

```python
df2=pd.merge(staff_df, student_df, how='outer', on=['First Name', 'Last Name'])
```

```python
print(len(student_df))
print(len(staff_df))
print(len(df2))
```

```python
frames=[student_df, staff_df, df2]
```

```python
pd.concat(frames)
```

```python
pd.concat(frames, keys=['student','staff','second'])
```
